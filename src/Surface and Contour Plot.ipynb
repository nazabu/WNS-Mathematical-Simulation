```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

def compute_R0(beta_d, gamma, beta_i, alpha, mu):
    R0 = (beta_d / gamma) + ((beta_i / gamma) * (alpha / mu))
    return R0

beta_d_range = np.linspace(0.01, 0.05, 100)
mu_range = np.linspace(0.01, 0.05, 100)

beta_d_mesh, mu_mesh = np.meshgrid(beta_d_range, mu_range)

R0_mesh = compute_R0(beta_d_mesh, mu_mesh, 0.5, 0.04, 0.05)


plt.figure(figsize=(8,6))
contour = plt.contourf(beta_d_mesh, mu_mesh, R0_mesh, levels=50, cmap="coolwarm")
plt.colorbar(contour, label="R0 Value")
plt.xlabel("Direct Transmission Rate (β_d)")
plt.ylabel("Host Mortality Rate (γ)")
plt.title("Contour Plot of R0 as a Function of β_d and γ")
plt.show()

# Ensure level values are strictly increasing
if R0_mesh.min() < 1 < R0_mesh.max():
    plt.contourf(beta_d_mesh, mu_mesh, R0_mesh, levels=np.linspace(R0_mesh.min(), 1 - 1e-6, 50), colors='blue', alpha=0.5)  # Below R0=1
    plt.contourf(beta_d_mesh, mu_mesh, R0_mesh, levels=np.linspace(1 + 1e-6, R0_mesh.max(), 50), colors='red', alpha=0.5)  # Above R0=1



fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(beta_d_mesh, mu_mesh, R0_mesh, cmap="viridis")

Z_plane = np.ones_like(R0_mesh)
ax.plot_surface(beta_d_mesh, mu_mesh, Z_plane, color='red', alpha=0.5)

ax.set_xlabel("β_d")
ax.set_ylabel("γ")
ax.set_zlabel("R0")
plt.title("3D Surface Plot of R0")
ax.view_init(elev=30., azim=35)
plt.show()
```
