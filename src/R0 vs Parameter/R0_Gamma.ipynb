```python
import numpy as np
import matplotlib.pyplot as plt

# Fixed parameters
alpha = 0.04   # Shedding rate of pathogen by infected hosts
mu = 0.05     # Pathogen decay rate
beta_d = 0.05  # Direct transmission rate
beta_i = 0.03  # Indirect transmission rate via environment

# Varying parameter (lower bound, higher bound, and resolution)
gamma_range = np.linspace(0.01, 0.5, 100)

R0_values = []

for gamma in gamma_range:
    R0 = (beta_d / gamma) + (beta_i / gamma) * (alpha / mu)
    R0_values.append(R0) # List of exact R0 values
```

```python
# Plot
plt.figure(figsize=(10, 6))
plt.plot(gamma_range, R0_values, label=r'$R_0$', color='purple')
plt.axhline(y=1, color='r', linestyle='--', label=r'$R_0 = 1$ (Threshold)')
plt.xlabel(r'$\gamma$ (Host Mortality Rate)', fontsize=12)
plt.ylabel(r'$R_0$ (Basic Reproduction Number)', fontsize=12)
plt.title(r'$R_0$ as a Function of $\gamma$', fontsize=14)
plt.legend()
plt.grid()
plt.show()

print([tuple((round(float(gamma), 3), round(float(R0), 3))) for gamma, R0 in zip(gamma_range, R0_values)])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Fixed parameters
alpha = 0.04
mu = 0.05
beta_d = 0.05
beta_i = 0.03

# Varying parameter (gamma range)
gamma_range = np.linspace(0.01, 0.5, 100)

# Compute R0 values
R0_values = [(beta_d / gamma) + (beta_i / gamma) * (alpha / mu) for gamma in gamma_range]

# Define a function for curve fitting (inverse relationship)
def inverse_model(gamma, a, b):
    return a / gamma + b

# Fit the curve to the data
params, _ = curve_fit(inverse_model, gamma_range, R0_values)

# Extract the best-fit parameters
a_fit, b_fit = params
print(f"Fitted Equation: R0 ≈ {a_fit:.4f} / γ + {b_fit:.4f}")

# Plot the original data and the fitted curve
plt.scatter(gamma_range, R0_values, label="Original Data", color="blue", s=10)
plt.plot(gamma_range, inverse_model(gamma_range, *params), label="Fitted Curve", color="red")
plt.xlabel("Gamma (γ)")
plt.ylabel("R0")
plt.legend()
plt.title("Fitting R0 vs Gamma")
plt.show()
```
