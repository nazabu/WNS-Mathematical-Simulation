```python
import numpy as np
import matplotlib.pyplot as plt

# Fixed parameters
gamma = 0.5  # Host mortality rate
alpha = 0.04  # Shedding rate of pathogen by infected hosts
mu = 0.05  # Pathogen decay rate
beta_i = 0.03  # Direct transmission rate via environment

# Varying parameter
beta_d_range = np.linspace(0.01, 0.5, 100)  # lower bound, higher bound, and resolution

R0_values = []

for beta_d in beta_d_range:
    R0 = (beta_d / gamma) + (beta_i / gamma) * (alpha / mu)
    R0_values.append(R0)  # List of exact R0 values
```

```python
# Plot
plt.figure(figsize=(10, 6))
plt.plot(beta_d_range, R0_values, label=r'$R_0$', color='b')
plt.axhline(y=1, color='r', linestyle='--', label=r'$R_0 = 1$ (Threshold)')
plt.xlabel(r'$β_d$ (Direct Transmission Rate)', fontsize=12)
plt.ylabel(r'$R_0$ (Basic Reproduction Number)', fontsize=12)
plt.title(r'$R_0$ as a Function of $β_d$', fontsize=14)
plt.legend()
plt.grid()
plt.show()

print([tuple((round(float(beta_d), 3), round(float(R0), 3))) for beta_d, R0 in zip(beta_d_range, R0_values)])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fixed parameters
gamma = 0.5
alpha = 0.04
mu = 0.05
beta_i = 0.03

# Varying parameter
beta_d_range = np.linspace(0.01, 0.5, 100)

# Compute R0 values
R0_values = [(beta_d / gamma) + (beta_i / gamma) * (alpha / mu) for beta_d in beta_d_range]

# Fit a linear model (degree = 1)
m, b = np.polyfit(beta_d_range, R0_values, 1)

# Print the best-fit equation
print(f"Fitted Equation: R0 ≈ {m:.4f} * β_d + {b:.4f}")

# Plot the original data and the fitted line
plt.scatter(beta_d_range, R0_values, label="Original Data", color="blue", s=10)
plt.plot(beta_d_range, m * beta_d_range + b, label="Fitted Line", color="red")
plt.xlabel("Direct Transmission Rate (β_d)")
plt.ylabel("R0")
plt.legend()
plt.title("Linear Fit for R0 vs β_d")
plt.show()
```
