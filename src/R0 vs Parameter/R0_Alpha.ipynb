```python
import numpy as np
import matplotlib.pyplot as plt

# Fixed parameters
gamma = 0.5 # Host mortality rate
mu = 0.05     # Pathogen decay rate
beta_d = 0.05  # Direct transmission rate
beta_i = 0.03  # Indirect transmission rate via environment


# Varying parameter (lower bound, higher bound, and resolution)
alpha_range = np.linspace(0.01, 0.5, 100)


R0_values = []

for alpha in alpha_range:
    R0 = (beta_d / gamma) + (beta_i / gamma) * (alpha / mu)
    R0_values.append(R0) # List of exact R0 values
```

```python
# Plot
plt.figure(figsize=(10, 6))
plt.plot(alpha_range, R0_values, label=r'$R_0$', color='turquoise')
plt.axhline(y=1, color='r', linestyle='--', label=r'$R_0 = 1$ (Threshold)')
plt.xlabel(r'$\alpha$ (Shedding Rate)', fontsize=12)
plt.ylabel(r'$R_0$ (Basic Reproduction Number)', fontsize=12)
plt.title(r'$R_0$ as a Function of $\alpha$', fontsize=14)
plt.legend()
plt.grid()
plt.show()

print([tuple((round(float(alpha), 3), round(float(R0), 3))) for alpha, R0 in zip(alpha_range, R0_values)])
```

```python
# Fit a linear model (degree = 1)
m, b = np.polyfit(alpha_range, R0_values, 1)

# Print the best-fit equation
print(f"Fitted Equation: R0 ≈ {m:.4f} * β_d + {b:.4f}")

# Plot the original data and the fitted line
plt.scatter(alpha_range, R0_values, label="Original Data", color="blue", s=10)
plt.plot(alpha_range, m * alpha_range + b, label="Fitted Line", color="red")
plt.xlabel(r'$\alpha$ (Shedding Rate)')
plt.ylabel("R0")
plt.legend()
plt.title(r'Linear fit $R_0$ as a Function of $\alpha$')
plt.show()
```
