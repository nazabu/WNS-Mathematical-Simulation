```python
import numpy as np
import matplotlib.pyplot as plt

# Fixed parameters
gamma = 0.5 # Host mortality rate
alpha = 0.04   # Shedding rate of pathogen by infected hosts
mu = 0.05     # Pathogen decay rate
beta_d = 0.03  # Indirect transmission rate via environment


# Varying parameter (lower bound, higher bound, and resolution)
beta_i_range = np.linspace(0.01, 0.5, 100)

R0_values = []

for beta_i in beta_i_range:
    R0 = (beta_d / gamma) + (beta_i / gamma) * (alpha / mu)
    R0_values.append(R0) # List of exact R0 values

```

```python
# Plot
plt.figure(figsize=(10, 6))
plt.plot(beta_i_range, R0_values, label=r'$R_0$', color='y')
plt.axhline(y=1, color='r', linestyle='--', label=r'$R_0 = 1$ (Threshold)')
plt.xlabel(r'$β_i$ (Indirect Transmission Rate)', fontsize=12)
plt.ylabel(r'$R_0$ (Basic Reproduction Number)', fontsize=12)
plt.title(r'$R_0$ as a Function of $β_i$', fontsize=14)
plt.legend()
plt.grid()
plt.show()

print([tuple((round(float(beta_i), 3), round(float(R0), 3))) for beta_i, R0 in zip(beta_i_range, R0_values)])
```

```python
# Fit a linear model (degree = 1)
m, b = np.polyfit(beta_i_range, R0_values, 1)

# Print the best-fit equation
print(f"Fitted Equation: R0 ≈ {m:.4f} * β_i + {b:.4f}")

# Plot the original data and the fitted line
plt.scatter(beta_i_range, R0_values, label="Original Data", color="blue", s=3)
plt.plot(beta_i_range, m * beta_i_range + b, label="Fitted Line", color="red")
plt.xlabel("Indirect Transmission Rate (β_i)")
plt.ylabel("R0")
plt.legend()
plt.title("Linear Fit for R0 vs β_i")
plt.show()
```
